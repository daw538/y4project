Pressure modes can only propagate within regions that are convectively unstable. Convection in a star will occur if a small region of gas is peturbed from its equilibrium position but this pertubation grows with time. This is described by the Schwarzschild Criterion defined thus
    $$ \nabla_R > \nabla_{ad} $$
where $\nabla_{ad}=\frac{2}{5}$ is a fully ionised gas, and $\nabla_R = \frac{d\ln T}{d\ln P}$. This can alternatively be expressed using the first adiabatic exponent $\Gamma_1$ where
    $$ \frac{d\ln\rho}{d\ln P} < \frac{1}{\Gamma_1}.$$
This shows us that the adiabatic exponent can be influenced by changes in certain factors, such as density or temperature.

It has been observed that rapid changes to the adiabatic exponent, caused by sharp changes in the interior stellar stratification,  can effect the p-mode oscillations.
