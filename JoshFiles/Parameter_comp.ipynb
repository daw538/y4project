```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pystan
```

```python
IDs = [1433593, 2018392, 2424934, 4469602, 5288607, 8872979, 9941141, 10614602, 10935853, 12366461]
pvalues = []
for i in IDs:
    pvalues.append(pd.read_csv('Samples/samples_' + str(i) + '.csv', delim_whitespace=False, header=0, index_col=0))

names = list(pvalues[0].columns.values)

fig, axs = plt.subplots(8, 1, figsize=(15, 20), sharey=False)

max_min = {
    "dnu_min" : 0,
    "dnu_max" : 0,
    "nmax_min" : 0,
    "nmax_max" : 0,
    "epsilon_min" : 0,
    "epsilon_max" : 0,
    "k_min" : 0,
    "k_max" : 0,
    "A_min" : 0,
    "A_max" : 0,
    "w_min" : 0,
    "w_max" : 0,
    "phi_min" : 0,
    "phi_max" : 0,
    "tau_min" : 0,
    "tau_max" : 0
}


for i in range(len(names)):
    mins = []
    maxs = []
    for j in pvalues:
        mins.append(min(j[names[i]]))
        maxs.append(max(j[names[i]]))
        axs[i].hist(j[names[i]])
        axs[i].set_title(names[i])
    minn = names[i] + '_min'
    maxn = names[i] + '_max'
    max_min[minn] = min(mins)
    max_min[maxn] = max(maxs)
fig.suptitle('Parameters')
```
